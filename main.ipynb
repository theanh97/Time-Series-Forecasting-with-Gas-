# Fetch AAPL stock & PLOT 

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# function to get the data for the stock
def get_stock_data(stock, start, end):
    """
    Downloads historical stock data for a given stock symbol within a specified date range.

    Parameters:
    - stock (str): The stock symbol to retrieve data for.
    - start (str): The start date of the data range in the format 'YYYY-MM-DD'.
    - end (str): The end date of the data range in the format 'YYYY-MM-DD'.

    Returns:
    - data (pandas.DataFrame): The historical stock data as a pandas DataFrame.
    """
    data = yf.download(stock, start, end)
    return data

# get the stock data 
ticker = 'AAPL'
start_date = '2018-01-01'
end_date = '2023-01-01'
df = get_stock_data(ticker, start_date, end_date)

# print the first 5 rows of the dataframe
print(df.head())
```

```python
""" 
    Plot "Adj Close" price
"""
# plot the adjusted close price
plt.figure(figsize=(12,8))
plt.plot(df['Adj Close'], label='Adj Close')
plt.title('Adjusted Close Price of ' + ticker)
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.show()
```
